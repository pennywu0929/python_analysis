```python
df = pd.read_excel('students.xlsx', index_col=0) # df代表dataframe，index_col=0代表第一個欄位是索引

df.head() #顯示前五筆資料   
```

```python
#透過欄選取
#單欄選取
df['國文'] #一維資料是series
```

```python
#多欄選取
df[['國文','英文']]#二維資料是dataframe，第一個括弧是所選的欄位名稱，第二個括弧是list(要顯示多個)
```

```python
#透過列選取
#loc->使用索引的名稱抓取
#iloc->使用索引的index值抓取
df.loc['王泓泰']
df.loc[['王泓泰','洪阿惠']]
```

```python
df.iloc[0]
df.iloc[[0,1]]
```

```python
#兩列以上
df.loc['王泓泰':'陳姵彬'] #抓取王泓泰到陳姵彬的資料
```

```python
df.iloc[0:3] #[:]選取序列，抓取第0列到第3列(不包含第三列)的資料
```

```python
#修改某一格資料數值
df.head()
df.loc['王泓泰','國文'] = 61 
df.head()
```

```python
#顯示前10筆資料
df[:10]
```

```python
#顯示後10筆資料
df[-10:]
```

```python
#加總所有數值
df['國文'] + df['英文'] + df['數學'] + df['自然'] + df['社會']
```

```python
#【axis=0】表述列，【axis=1】表述行
stu_sum = df.sum(axis=1)
stu_sum
```

```python
#每人的平均分數
stu_mean = stu_sum / 5.0
stu_mean
```

```python
df['總分'] = stu_sum
df['平均'] = stu_mean
df
```
